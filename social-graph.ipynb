```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.read_edgelist("Dataset/facebook_combined/facebook_combined.txt")
```

```python
layout = nx.spring_layout(G, iterations=20, seed=111)
options = {"node_size":10, "with_labels":False, "width":0.15}
```

```python
plt.figure(figsize=(20, 10))
nx.draw(G, pos=layout, **options)
plt.title("Graph Interaksi Jaringan Sosial")
plt.show()
```

## Closeness Centrality 

```python
cc = nx.centrality.closeness_centrality(G)
```

```python
closeness_centrality = (sorted(cc.items(), key=lambda item:item[1], reverse=True))[:10]
```

```python
top_nodes = (sorted(cc.items(), key=lambda item:item[1], reverse=True))[:10]
neighbors_of_top_centrality_nodes = {}
for node, centrality in top_nodes:
    neighbors = list(G.neighbors(node))
    neighbors_of_top_centrality_nodes[node] = neighbors[:9]
```

```python
node = [item[0] for item in closeness_centrality]
cent_values = [item[1] for item in closeness_centrality]

# Create a bar chart with nodes on the y-axis and values on the x-axis
plt.figure(figsize=(20, 10))
plt.barh(node, cent_values)  # Use barh to create a horizontal bar chart

# Add labels and title
plt.xlabel('Closeness Centrality Values')
plt.ylabel('Nodes')
plt.title('High Centrality Nodes Using Closeness Centrality ')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
neighbors_of_top_centrality_nodes
```

```python
plt.figure(figsize=(20, 10))
plt.hist(cc.values(), bins=60)
plt.show()
```

```python
node_size = [v * 50 for v in cc.values()]
plt.figure(figsize=(20, 10))
nx.draw_networkx(G, layout, node_size=node_size, with_labels=False, width=0.15)
plt.title("Grafik Interaksi Setelah Perhitungan Centrality")
plt.show()
```

## Betweenness Centrality

```python
cb = nx.centrality.betweenness_centrality(G)
```

```python
(sorted(cb.items(), key=lambda item:item[1], reverse=True))[:10]
```

```python
plt.hist(cb.values(), bins=60)
```

```python
betweenness_centrality = (sorted(cb.items(), key=lambda item:item[1], reverse=True))[:10]

node = [item[0] for item in betweenness_centrality]
cent_values = [item[1] for item in betweenness_centrality]

# Create a bar chart with nodes on the y-axis and values on the x-axis
plt.figure(figsize=(20, 10))
plt.barh(node, cent_values)  # Use barh to create a horizontal bar chart

# Add labels and title
plt.xlabel('Betweenness Centrality Values')
plt.ylabel('Nodes')
plt.title('High Centrality Nodes Using Betweenness Centrality ')

# Show the plot
plt.tight_layout()
plt.show()
```
