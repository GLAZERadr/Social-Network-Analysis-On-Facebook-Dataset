```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.read_edgelist("Dataset/facebook_combined/facebook_combined.txt")
```

```python
layout = nx.spring_layout(G, iterations=10, seed=70)
options = {"node_size":10, "with_labels":False, "width":0.15}
```

```python
plt.figure(figsize=(20, 10))
nx.draw(G, pos=layout, **options)
plt.title("Graph Interaksi Jaringan Sosial")
plt.show()
```

## Closeness Centrality 

```python
cc = nx.centrality.closeness_centrality(G)
```

```python
(sorted(cc.items(), key=lambda item:item[1], reverse=True))[:10]
```

```python
top_nodes = (sorted(cc.items(), key=lambda item:item[1], reverse=True))[:10]
neighbors_of_top_centrality_nodes = {}
for node, centrality in top_nodes:
    neighbors = list(G.neighbors(node))
    neighbors_of_top_centrality_nodes[node] = neighbors[:9]
```

```python
neighbors_of_top_centrality_nodes
```

```python
plt.hist(cc.values(), bins=60)
```

```python
node_size = [v * 50 for v in cc.values()]
plt.figure(figsize=(20, 10))
nx.draw_networkx(G, layout, node_size=node_size, with_labels=False, width=0.15)
plt.title("Grafik Interaksi Setelah Perhitungan Centrality")
plt.show()
```
